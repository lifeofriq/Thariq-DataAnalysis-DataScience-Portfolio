#Data Preprocessing

###One Hot Encoding

First Step, we need First, we need to convert 'RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm' columns from categorical variables to binary variables

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

The next step is we going to convert "No" and "Yes Value to 0 and 1 (Binary Value)

```python
df_sydney_processed.replace(['No', 'Yes'], [0, 1], inplace=True)
```

#Splitting the data into train and test set

Setting up our X values and y or target variable.

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
X = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
y = df_sydney_processed['RainTomorrow']
```

#### Q1) Use the `train_test_split` function to split the `features` and `Y` dataframes with a `test_size` of `0.2` and the `random_state` set to `10`.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
print(f"x_train shape: {x_train.shape}")
print(f"x_test shape: {x_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

#Linear Regression

#### Q2) Create and train a Linear Regression model called LinearReg using the training data (`x_train`, `y_train`).

```python
from sklearn import linear_model
LinearReg = linear_model.LinearRegression()
train_x = np.array(x_train)
train_y = np.array(y_train)
LinearReg.fit(train_x, train_y)

print ('Coefficients: ', LinearReg.coef_)
```

#### Q3) Now use the `predict` method on the testing data (`x_test`) and save it to the array `predictions`.

```python
predictions = LinearReg.predict(x_test)

print("Predictions ", predictions)
```

#### Q4) Using the `predictions` and the `y_test` dataframe calculate the value for each metric using the appropriate function.

```python
from sklearn.metrics import r2_score

test_x = np.array(x_test)
test_y = np.array(y_test)
test_y_ = LinearReg.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

#### Q5) Show the MAE, MSE, and R2 in a tabular format using data frame for the linear model.

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate metrics
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

# Create a DataFrame for the metrics
metrics_df = pd.DataFrame({
    "Metric": ["Mean Absolute Error (MAE)", "Mean Squared Error (MSE)", "RÂ²-score"],
    "Value": [mae, mse, r2]
})

# Display the DataFrame
print(metrics_df)
```
